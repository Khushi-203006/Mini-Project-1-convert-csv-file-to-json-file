<h1 style="text-align:center;">Retail Dataset Conversion Project</h1>

<h2>Project Overview</h2>

<p>
This project converts raw retail dataset files into structured <b>JSON format</b> using 
<b>Python</b> and <b>Pandas</b>. The purpose of this project is to transform unstructured 
or semi-structured text files into a well-structured format that can be easily used in 
data processing systems.
</p>

<p>
The pipeline reads multiple datasets such as <b>orders, products, customers, categories,
departments,</b> and <b>order_items</b>. These datasets are originally stored as text files
inside the <b>retail_db</b> directory.
</p>

<p>
A configuration file called <b>schemas.json</b> is used to dynamically assign the column
names to each dataset. This ensures that the data is read correctly and the column order
is maintained properly.
</p>

<p>
After loading the data using Pandas, the program converts the data into <b>JSON format</b>
and saves it into a separate output directory called <b>retail_db_json</b>.
</p>

<p>
This approach makes the dataset easier to use in <b>data pipelines, big data systems,
analytics platforms, and APIs</b>.
</p>

<h2>Key Functions in the Project</h2>

<ul>
<li><b>get_column_name()</b> – Extracts column names from the schema file.</li>
<li><b>read_csv()</b> – Reads dataset files and converts them into a Pandas DataFrame.</li>
<li><b>to_json()</b> – Converts the DataFrame into JSON format.</li>
<li><b>file_converter()</b> – Processes dataset files and applies the conversion.</li>
<li><b>process_file()</b> – Controls the complete data processing pipeline.</li>
</ul>
